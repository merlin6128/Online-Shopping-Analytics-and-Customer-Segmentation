### Step 1: Import the following library

### Step 2: Download data from UCI
Head on to the <a href="https://archive.ics.uci.edu/ml/datasets/Online+Retail+II">UCI Machine Learning Repository</a> and download the dataset from <a href="https://archive.ics.uci.edu/ml/machine-learning-databases/00502/online_retail_II.xlsx">here</a>. 

### Step 3: Read Excel file into a DataFrame
There are two ways to proceed from here:
1. Store the two sheets in two separate variables first, then combine them into one
2. Do 1 but in a single line of code

```python
# Step 3a: Read Excel sheets into two separate DataFrames
xls = pd.ExcelFile('online_retail_II.xlsx')
df1 = pd.read_excel(xls, 'Year 2009-2010')
df2 = pd.read_excel(xls, 'Year 2010-2011')
```

```python
# Step 3b: Concatenate two DataFrames into a single DataFrame and reset the index
frames = [df1, df2]
  
df_concat = pd.concat(frames)
display(df_concat)
```

### Step 4: Set 'InvoiceDate' as DataFrame's index
Used 'InvoiceDate' as the index

```python
# Step 4: Set InvoiceDate as index
df_inv = df_concat.set_index('InvoiceDate')
```

### Step 5: Take a quick look at numerical data using .describe()
Let's see if there is anything right/wrong with the data. Use .describe to look the summary of the data.

```python
# Step 5: Use the .describe method on the DataFrame
data = df_inv.copy()
data.describe()
```

### Step 6: Check rows that contain negative quantity
The minimum is not 0. It's a negative value, and a huge one at that. Not possible!

Subset the DataFrame to get rows that contain only values that are less than 0. 

```python
# Step 6a: Filter rows based on "Quantity" that's less than 0
data.loc[data['Quantity'] < 0]
```

```python
# Step 6b: Filter rows based on "Price" that's less than 0
data.loc[data['Price'] < 0]
```

### Step 8: Remove negative quantities of Quantity and Price
There are 22,950 rows that contain negative Quantity, and 5 rows that contain negative Price. 

Remove those to have only positive numbers in the DataFrame. 

Expect a final DataFrame with <strong>1,041,671 rows</strong> with 7 columns

```python
# Step 8: Remove rows that contain negative Quantity and Price
data = data.loc[(data["Quantity"] > 0) & (data["Price"] > 0)]
data
```

### Step 9: Export the DataFrame as CSV
Removed 22k rows

Export the DataFrame as a CSV for Exploratory Data Analysis in Part II.

```python
# Step 9: Export the DataFrame as CSV
data.to_csv ('DataFrame.csv')
```
